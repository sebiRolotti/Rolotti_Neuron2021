```python
def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)

def plot_inset(df, labels, colors, ylabel, save_path, yticks=None):

    fig = plt.figure(figsize=(0.5, 0.5), dpi=200)
    ax = fig.add_subplot(111)

    sns.boxplot(y='value', x='label', order=labels, data=df, ax=ax,
                whis=[5, 95], showcaps=True, showfliers=False, dodge=True)

    for artist, edgecolor, facecolor, ls in zip(ax.artists, ['r', 'r', '0.7'], ['r', 'w', '0.7'], ['-', '--', '-', '-']):
        plt.setp(artist, edgecolor=edgecolor, facecolor=facecolor, ls=ls, lw=0.75)

    plt.setp(ax.lines, color='k', linewidth=0.75)

    ax.legend().set_visible(False)

    clean(ax, offset=2)
    sns.despine(ax=ax, right=True, top=True, bottom=True)
    
    ax.set_ylabel('')
    ax.set_xlabel('')
    if yticks is not None:
        ax.set_yticks(yticks)
    else:
        ax.set_yticks(ax.get_yticks()[::2])
    ax.set_xticks([])
    
    ax.set_aspect(2 / ax.get_data_ratio())
    
#     fig.savefig(save_path, bbox_inches='tight')
```

```python
def plot_cdf(df, labels, colors, linestyles, ylabel, save_path):

    fig = plt.figure(figsize=(1, 1.5), dpi=200)
    ax = fig.add_subplot(111)

    for label, color, ls in zip(labels, colors, linestyles):

        subdf = df.loc[(df['label'] == label)]
        bins, cdf_vals = calc_cdf(ax, subdf['value'], color=color, ls=ls, lw=1)
        ax.vlines(ip.get_med(cdf_vals, bins), 0, 0.5, linestyles=ls, color=color, alpha=0.5, lw=1)

    if np.nanmax(df.loc[(df['label'] == labels[-1])]['value']) > 2* np.nanmax(df.loc[(df['label'] == labels[0])]['value']):
        ax.set_xlim([ax.get_xlim()[0], np.nanpercentile(df.loc[(df['label'] == labels[-1])]['value'], 99)])

    clean(ax)
    ax.set_ylabel('Cumulative Fraction')
    ax.set_xlabel(ylabel)
    
    fig.savefig(save_path, bbox_inches='tight')
```

```python
def event_rate(grp, roi_filter=None, running=True):

    data_list = []
    for expt in grp:

        tid = expt.trial_id

        rois = expt.rois(roi_filter=roi_filter)
        spikes = expt.spikes(roi_filter=roi_filter, binary=True)

        if running:
            interval = np.where(expt.velocity()[0] > 1)[0]
        else:
            interval = np.where(expt.velocity()[0] <= 1)[0]

        frs = np.nansum(spikes[:, interval], axis=1) / (len(interval) * expt.frame_period())

        for roi, fr in zip(rois, frs):

            data_list.append({'expt_id': tid,
                              'roi': roi.label,
                              'value': fr})

    return pd.DataFrame(data_list)
```

## Recurrence Probability

```python
def recurrence_probability(grp, roi_filter=None, session2='control_24h'):
    """Calculate Fraction of Place cells still place cells on the next day

    pc_only governs pf criteria for roi filtering.
    """
    pc_kwargs = grp.args

    data_list = []

    for expt in grp:

        trial_id = expt.trial_id
        mouse_name = expt.parent.mouse_name

        sess2 = expt._get_session(session2)

        pcgrp = pcExperimentGroup([sess2], **pc_kwargs)

        pfs = pcgrp.pfs(roi_filter=roi_filter)[sess2]
        rois = sess2.rois(roi_filter=roi_filter)

        n_rois = len(rois)
        n_pfs = len([x for x in pfs if x])
        
        try:
            rprob = float(n_pfs) / n_rois
        except ZeroDivisionError:
            rprob = np.nan

        data_list.append({
            'expt_id': trial_id,
            'mouse_name': mouse_name,
            'n_rois': n_rois,
            'n_recurring': n_pfs,
            'recurrence_prob': rprob
        })

    return pd.DataFrame(data_list)
```

```python
nostim_recurrence = recurrence_probability(grp, roi_filter=nostim_pc_filter)
ind_recurrence = recurrence_probability(grp, roi_filter=ind_filter)
```
